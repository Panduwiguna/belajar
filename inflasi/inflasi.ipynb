```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn untuk gaya plot yang lebih baik jika diinginkan

# --- Konfigurasi Awal (Pastikan Sesuai Dataset Anda) ---
NAMA_FILE_CSV = 'indonesia.csv'
KOLOM_TAHUN = 'TAHUN'
KOLOM_INFLASI = 'INFLASI'
KOLOM_PDB = 'GDP' # Tambahkan kolom PDB

print("--- Membuat Diagram Batang Tingkat Inflasi dan PDB ---")

# --- 1. Muat Dataset ---
try:
    df = pd.read_csv(NAMA_FILE_CSV)
    print(f"\nDataset '{NAMA_FILE_CSV}' berhasil dimuat!")
except FileNotFoundError:
    print(f"ERROR: File '{NAMA_FILE_CSV}' tidak ditemukan.")
    print("Pastikan file tersebut berada di direktori yang sama dengan notebook Anda.")
    exit()

# Cek keberadaan kolom
if KOLOM_TAHUN not in df.columns or KOLOM_INFLASI not in df.columns or KOLOM_PDB not in df.columns:
    print(f"ERROR: Kolom '{KOLOM_TAHUN}', '{KOLOM_INFLASI}', atau '{KOLOM_PDB}' tidak ditemukan.")
    print(f"Kolom yang tersedia: {df.columns.tolist()}")
    exit()

# --- 2. Buat Diagram Batang untuk Inflasi ---
plt.figure(figsize=(14, 7)) # Ukuran figure (lebar, tinggi) dalam inci
plt.bar(df[KOLOM_TAHUN], df[KOLOM_INFLASI], color='skyblue') # Menggunakan plt.bar

# --- 3. Tambahkan Judul dan Label Sumbu untuk Inflasi ---
plt.title("Tingkat Inflasi Tahunan di Indonesia (1980-2022)", fontsize=16)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Tingkat Inflasi (%)", fontsize=12)

# --- 4. Penyesuaian Tampilan untuk Inflasi ---
plt.xticks(rotation=90, fontsize=10) # Memutar label tahun agar tidak tumpang tindih
plt.grid(axis='y', linestyle='--') # Menampilkan grid hanya di sumbu y
plt.tight_layout() # Menyesuaikan layout agar semua elemen terlihat rapi

# --- 5. Tampilkan Grafik Inflasi ---
plt.show()

print("\n--- Diagram Batang Tingkat Inflasi Selesai Dibuat ---")

# --- 6. Buat Diagram Batang untuk PDB ---
plt.figure(figsize=(14, 7)) # Ukuran figure (lebar, tinggi) dalam inci
plt.bar(df[KOLOM_TAHUN], df[KOLOM_PDB], color='lightcoral') # Menggunakan plt.bar, warna berbeda

# --- 7. Tambahkan Judul dan Label Sumbu untuk PDB ---
plt.title("Pertumbuhan Produk Domestik Bruto (PDB) Tahunan di Indonesia (1980-2022)", fontsize=16)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Pertumbuhan PDB (%)", fontsize=12)

# --- 8. Penyesuaian Tampilan untuk PDB ---
plt.xticks(rotation=90, fontsize=10) # Memutar label tahun agar tidak tumpang tindih
plt.grid(axis='y', linestyle='--') # Menampilkan grid hanya di sumbu y
plt.tight_layout() # Menyesuaikan layout agar semua elemen terlihat rapi

# --- 9. Tampilkan Grafik PDB ---
plt.show()

print("\n--- Diagram Batang Pertumbuhan PDB Selesai Dibuat ---")
```

```python
# Import necessary libraries at the top
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

print("\n--- Melakukan Regresi Linier Sederhana (GDP -> Inflasi) ---")

X = df[['GDP']].values
Y = df['INFLASI'].values

# Initialize and Train the Linear Regression Model
model = LinearRegression()
model.fit(X, Y)

# Get Coefficient and Intercept
intercept = model.intercept_
coefficient = model.coef_[0]

print(f"Persamaan Regresi: Inflasi = {intercept:.2f} + {coefficient:.2f} * GDP")
print(f"Intercept (b0): {intercept:.2f}")
print(f"Koefisien GDP (b1): {coefficient:.2f}")

# Interpretation (for your report)
print("\nInterpretasi Koefisien:")
print(f"- Intercept ({intercept:.2f}): Perkiraan tingkat inflasi jika pertumbuhan GDP adalah 0%.")
print(f"- Koefisien GDP ({coefficient:.2f}): Setiap peningkatan 1% pada Pertumbuhan GDP, diperkirakan akan mengakibatkan perubahan inflasi sebesar {coefficient:.2f}%.")

# --- Optional: Add a plot for the regression line ---
plt.figure(figsize=(10, 7))
sns.scatterplot(x=df['GDP'], y=df['INFLASI'], color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', linestyle='--', label='Regression Line')
plt.title('Linear Regression: GDP vs Inflation')
plt.xlabel('GDP Growth (%)')
plt.ylabel('Inflation Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python


# --- Plot untuk Inflasi ---
mean_inflasi = df['INFLASI'].mean()
std_inflasi = df['INFLASI'].std()


plt.errorbar(['Inflasi'], [mean_inflasi], yerr=[std_inflasi], fmt='o', capsize=5, label=f'Mean: {mean_inflasi:.2f} +/- Std Dev: {std_inflasi:.2f}')
plt.title('Mean dan Standard Deviasi Inflasi')
plt.ylabel('Tingkat Inflasi (%)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7) # Menambahkan grid pada sumbu Y
plt.show()

# --- Plot untuk GDP ---
mean_gdp = df['GDP'].mean()
std_gdp = df['GDP'].std()


plt.errorbar(['GDP'], [mean_gdp], yerr=[std_gdp], fmt='o', capsize=5, label=f'Mean: {mean_gdp:.2f} +/- Std Dev: {std_gdp:.2f}')
plt.title('Mean dan Standard Deviasi GDP')
plt.ylabel('Pertumbuhan GDP (%)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7) # Menambahkan grid pada sumbu Y
plt.show()
```
