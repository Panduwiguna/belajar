```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn untuk gaya plot yang lebih baik jika diinginkan

FILE_CSV = 'indonesia.csv'
KOLOM_TAHUN = 'TAHUN'
KOLOM_INFLASI = 'INFLASI'
KOLOM_PDB = 'GDP' # Tambahkan kolom PDB

try:
    df = pd.read_csv(FILE_CSV)
    print(f"\nDataset '{FILE_CSV}' berhasil dimuat!")
except FileNotFoundError:
    print(f"ERROR: File '{FILE_CSV}' tidak ditemukan.")
    print("Pastikan file tersebut berada di direktori yang sama dengan notebook Anda.")
    exit()


if KOLOM_TAHUN not in df.columns or KOLOM_INFLASI not in df.columns or KOLOM_PDB not in df.columns:
    print(f"ERROR: Kolom '{KOLOM_TAHUN}', '{KOLOM_INFLASI}', atau '{KOLOM_PDB}' tidak ditemukan.")
    print(f"Kolom yang tersedia: {df.columns.tolist()}")
    exit()


plt.figure(figsize=(14, 7))
plt.bar(df[KOLOM_TAHUN], df[KOLOM_INFLASI], color='skyblue') 

plt.title("Tingkat Inflasi di Indonesia (1980-2022)", fontsize=16)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Tingkat Inflasi (%)", fontsize=12)

plt.xticks(rotation=90, fontsize=10) 
plt.grid(axis='y', linestyle='--') 
plt.tight_layout() 
plt.show()



plt.figure(figsize=(14,7)) 
plt.bar(df[KOLOM_TAHUN], df[KOLOM_PDB], color='lightcoral')
plt.title("Pertumbuhan PDB di Indonesia (1980-2022)", fontsize=16)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Pertumbuhan PDB (%)", fontsize=12)

plt.xticks(rotation=90, fontsize=10) 
plt.grid(axis='y', linestyle='--') 
plt.tight_layout() 
plt.show()
```

```python
sns.regplot(x='INFLASI', y='GDP', data=df, scatter_kws={'alpha':0.6})
plt.title('Hubungan Inflasi terhadap PDB (Produk Domestik Bruto)')
plt.xlabel('Tingkat Inflasi (%)')
plt.ylabel('PDB')
plt.grid(True)
plt.show()
```

```python
plt.boxplot(df["INFLASI"], vert=False)
plt.title("Box Plot Distribusi Inflasi")
plt.xlabel("Tingkat Inflasi (%)")
plt.show()

plt.boxplot(df["GDP"], vert=False)
plt.title("Box Plot Distribusi GDP")
plt.xlabel("Pertumbuhan GDP (%)")
plt.show()
```

```python

mean_inflasi = df['INFLASI'].mean()
std_inflasi = df['INFLASI'].std()


plt.errorbar(['Inflasi'], [mean_inflasi], yerr=[std_inflasi], fmt='o', capsize=5, label=f'Mean: {mean_inflasi:.2f} +/- Std Dev: {std_inflasi:.2f}')
plt.title('Mean dan Standard Deviasi Inflasi')
plt.ylabel('Tingkat Inflasi (%)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


mean_gdp = df['GDP'].mean()
std_gdp = df['GDP'].std()
plt.errorbar(['GDP'], [mean_gdp], yerr=[std_gdp], fmt='o', capsize=5, label=f'Mean: {mean_gdp:.2f} +/- Std Dev: {std_gdp:.2f}')
plt.title('Mean dan Standard Deviasi GDP')
plt.ylabel('Pertumbuhan GDP (%)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show()
```
