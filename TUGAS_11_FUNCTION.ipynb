```python
def reverse_string(s): # Fungsi pembantu untuk membalik sebuah string.
    reversed_s = ""
    for char in s:
        reversed_s = char + reversed_s
    return reversed_s

   
def reverse_per_kata(kalimat): #Menerima satu string kalimat, lalu membalik setiap kaa tanpa mengubah urutan kata
    if not isinstance(kalimat, str):
        raise TypeError("Input harus berupa string.") #ini pemberitahuan kalo inputmya harus string

    kata_kata = kalimat.split(' ')
    kata_terbalik = []

    for kata in kata_kata:
        kata_terbalik.append(reverse_string(kata))

    return ' '.join(kata_terbalik)

def urutkan_kalimat(kalimat, urutan):#Menerima string kalimat dan list berisi angka (posisi baru dari kata-kata),lalu menghasilkan kalimat baru yang kata-katanya diurutkan sesuai indeks dalam urutan.Catatan: Angka di urutan dimulai dari 1 (bukan 0).
    if not isinstance(kalimat, str):
        raise TypeError("Argumen 'kalimat' harus berupa string.")
    if not isinstance(urutan, list) or not all(isinstance(i, int) for i in urutan):
        raise TypeError("Argumen 'urutan' harus berupa list berisi angka (integer).")

    kata_asli = kalimat.split(' ')
    jumlah_kata = len(kata_asli)

    # Validasi panjang list urutan
    if len(urutan) != jumlah_kata:
        raise ValueError("Jumlah angka dalam 'urutan' harus sama dengan jumlah kata dalam 'kalimat'.")

    # Validasi nilai dalam list urutan
    if not all(1 <= idx <= jumlah_kata for idx in urutan):
        raise ValueError(f"Angka dalam 'urutan' harus antara 1 dan {jumlah_kata}.")
    if len(set(urutan)) != jumlah_kata: # Memeriksa duplikasi atau angka yang terlewat
        raise ValueError("Angka dalam 'urutan' harus unik dan mencakup semua posisi kata.")

    # Buat list baru untuk menampung kata-kata yang sudah diurutkan
    # Inisialisasi dengan None atau string kosong sebanyak jumlah kata
    kata_terurut = [None] * jumlah_kata

    for i, posisi_baru in enumerate(urutan):
        # Ingat: posisi_baru dimulai dari 1, sedangkan indeks list dimulai dari 0
        # Jadi, posisi_baru - 1 adalah indeks yang benar untuk menempatkan kata
        kata_terurut[posisi_baru - 1] = kata_asli[i]

    return ' '.join(kata_terurut)

#Bagian Pengujian
if __name__ == "__main__":
    # Tugas 1: Mereverse setiap kata dalam kalimat
    print("--- Tugas 1: Mereverse Setiap Kata dalam Kalimat ---")
    input_kalimat_1 = "AKU CINTA KAMU"
    output_kalimat_1 = reverse_per_kata(input_kalimat_1)
    print(f"Input: \"{input_kalimat_1}\"")
    print(f"Output: \"{output_kalimat_1}\"")
    print() # Baris kosong untuk spasi
    print("-" * 50)

    input_kalimat_2 = "PYTHON ITU MUDAH"
    output_kalimat_2 = reverse_per_kata(input_kalimat_2)
    print(f"Input: \"{input_kalimat_2}\"")
    print(f"Output: \"{output_kalimat_2}\"")
    print() # Baris kosong untuk spasi
    print("-" * 50)

    input_kalimat_3 = "PROGRAMMING ITU MENYENANGKAN SEKALI"
    output_kalimat_3 = reverse_per_kata(input_kalimat_3)
    print(f"Input: \"{input_kalimat_3}\"")
    print(f"Output: \"{output_kalimat_3}\"")
    print() # Baris kosong untuk spasi
    print("-" * 50)

    # --- Tugas 2: Mengurutkan kata berdasarkan indeks list ---
    print("\n--- Tugas 2: Mengurutkan Kata Berdasarkan Indeks List ---")
    input_kalimat_4 = "HARI INI SEDANG BELAJAR PYTHON"
    input_urutan_4 = [5, 1, 4, 3, 2] # PYTHON HARI BELAJAR SEDANG INI
    output_kalimat_4 = urutkan_kalimat(input_kalimat_4, input_urutan_4)
    print(f"Input kalimat: \"{input_kalimat_4}\"")
    print(f"Input urutan: {input_urutan_4}")
    print(f"Output: \"{output_kalimat_4}\"")
    print() # Baris kosong untuk spasi
    print("-" * 50)

    input_kalimat_5 = "SAYA BELAJAR PYTHON SANGAT SERU"
    input_urutan_5 = [3, 2, 1, 5, 4] # PYTHON BELAJAR SAYA SERU SANGAT
    output_kalimat_5 = urutkan_kalimat(input_kalimat_5, input_urutan_5)
    print(f"Input kalimat: \"{input_kalimat_5}\"")
    print(f"Input urutan: {input_urutan_5}")
    print(f"Output: \"{output_kalimat_5}\"")
    print() # Baris kosong untuk spasi
    print("-" * 50)

    # Contoh dengan error handling
    print("\n--- Tugas 2: Contoh Error Handling ---")
    try:
        urutkan_kalimat("TIGA KATA INI", [1, 2]) # Urutan tidak sama dengan jumlah kata
    except ValueError as e:
        print(f"Error: {e}")
    print("-" * 50)

    try:
        urutkan_kalimat("TIGA KATA INI", [1, 2, 4]) # Angka di luar rentang
    except ValueError as e:
        print(f"Error: {e}")
    print("-" * 50)

    try:
        urutkan_kalimat("TIGA KATA INI", [1, 1, 3]) # Angka duplikat
    except ValueError as e:
        print(f"Error: {e}")
    print("-" * 50)

    try:
        urutkan_kalimat("TIGA KATA INI", "salah") # Tipe urutan salah
    except TypeError as e:
        print(f"Error: {e}")
    print("-" * 50)
```
