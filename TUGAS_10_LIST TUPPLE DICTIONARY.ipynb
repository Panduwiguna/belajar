```python
##### data_mahasiswa = {}

def tambah_mahasiswa():
    """Menambahkan data mahasiswa baru ke dalam sistem."""
    nim = input("Masukkan NIM: ")
    if nim in data_mahasiswa:
        print("Mahasiswa dengan NIM tersebut sudah ada.")
        return

    nama = input("Masukkan nama mahasiswa: ")
    jurusan = input("Masukkan jurusan: ")
    while True:
        try:
            ipk = float(input("Masukkan IPK mahasiswa: "))
            if not (0.0 <= ipk <= 4.0):
                print("IPK harus antara 0 sampai 4.")
                continue
            break
        except ValueError:
            print("Input IPK tidak benar, masukkan ulang.")

    # Menyimpan data mahasiswa dalam tuple (Nama, Jurusan, IPK)
    data_mahasiswa[nim] = (nama, jurusan, ipk)
    print(f"Data mahasiswa dengan NIM {nim} berhasil ditambahkan.")


def tampilkan_seluruh_mahasiswa():
    """Menampilkan seluruh data mahasiswa yang tersimpan."""
    if not data_mahasiswa:
        print("Belum ada data mahasiswa yang tersimpan.")
        return

    print("\n== SELURUH DATA MAHASISWA ==")
    for nim, data in data_mahasiswa.items():
        nama, jurusan, ipk = data
        print(f"NIM: {nim}")
        print(f"Nama: {nama}")
        print(f"Jurusan: {jurusan}")
        print(f"IPK: {ipk:.2f}")
        print("-" * 20)


def cari_mahasiswa():
    """Mencari data mahasiswa berdasarkan NIM."""
    nim = input("Masukkan NIM mahasiswa yang dicari: ")
    if nim in data_mahasiswa:
        nama, jurusan, ipk = data_mahasiswa[nim]
        print(f"\n== DATA MAHASISWA DITEMUKAN ==")
        print(f"NIM: {nim}")
        print(f"Nama: {nama}")
        print(f"Jurusan: {jurusan}")
        print(f"IPK: {ipk:.2f}")
    else:
        print("Mahasiswa dengan NIM tersebut tidak ditemukan.")


def update_ipk_mahasiswa():
    """Mengupdate data IPK mahasiswa berdasarkan NIM."""
    nim = input("Masukkan NIM mahasiswa yang akan diupdate IPK-nya: ")
    if nim in data_mahasiswa:
        nama, jurusan, _ = data_mahasiswa[nim]
        while True:
            try:
                new_ipk = float(input(f"Masukkan IPK baru untuk {nama} (NIM: {nim}): "))
                if not (0.0 <= new_ipk <= 4.0):
                    print("IPK harus antara 0 sampai 4.")
                    continue
                break
            except ValueError:
                print("Input IPK tidak valid, coba masukkan angka.")

        # UPDATE DATA IPK DENGAN MEMBUAT TUPLE BARU
        data_mahasiswa[nim] = (nama, jurusan, new_ipk)
        print(f"IPK mahasiswa dengan NIM {nim} berhasil diupdate menjadi {new_ipk:.2f}.")
    else:
        print("Mahasiswa dengan NIM tersebut tidak ditemukan.")

def hapus_mahasiswa():
    """Menghapus data mahasiswa berdasarkan NIM."""
    nim = input("Masukkan NIM mahasiswa yang ingin dihapus: ")
    if nim in data_mahasiswa:
        del data_mahasiswa[nim]
        print(f"Data mahasiswa dengan NIM {nim} berhasil dihapus.")
    else:
        print("Mahasiswa dengan NIM tersebut tidak ditemukan.")


def tampilkan_menu():
    """Menampilkan menu interaktif kepada pengguna."""
    print("\n--- Menu ---")
    print("1. Tambah Mahasiswa")
    print("2. Tampilkan seluruh mahasiswa")
    print("3. Cari Mahasiswa")
    print("4. Update IPK Mahasiswa")
    print("5. Hapus Mahasiswa")
    print("6. Keluar") 

def main():
    """Fungsi utama program untuk menjalankan menu interaktif."""
    while True:
        tampilkan_menu()
        pilihan = input("Pilihan Anda: ")

        if pilihan == '1':
            tambah_mahasiswa()
        elif pilihan == '2':
            tampilkan_seluruh_mahasiswa()
        elif pilihan == '3':
            cari_mahasiswa()
        elif pilihan == '4':
            update_ipk_mahasiswa()
        elif pilihan == '5':
            hapus_mahasiswa()
        else:
            print("Pilihan tidak valid, silakan coba lagi ya :)")

if __name__ == "__main__":
    main()
```
