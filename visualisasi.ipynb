```python
#MAtplotlib, SeaBorn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
bulan = ['Januari','Februari','Maret','April','Mei','Juni','Juli','Agustus','September','Oktober','November','Desember']
Minggu_1 = np.random.randint(200000,700001,12)
Minggu_2 = np.random.randint(200000,700001,12)
Minggu_3 = np.random.randint(200000,700001,12)
Minggu_4 = np.random.randint(200000,700001,12)

data = {
    'Bulan' : bulan,
    'Minggu1' : Minggu_1,
    'Minggu2' : Minggu_2,
    'Minggu3' : Minggu_3,
    'Minggu4' : Minggu_4,
}
data['Bulan']
df = pd.DataFrame(data)
df['Total_Pengeluaran'] = df[['Minggu1','Minggu2','Minggu3','Minggu4']].sum(axis=1)
df
```

```python
#DIAGRAM BAR
#BUAT CANVAS
plt.figure(figsize=(12,6)) #width, height -> inch
plt.bar(df['bulan'],df['Total_Pengeluaran'], color='skyblue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.title("Total Pengeluaran mahasiswa perbulan")
plt.ylabel("rupiah")
plt.grid(axis='y')
plt.show()
```

```python
#tren pengeluaran perminggu dalam satu tahun
plt.figure(figsize=(10,5))

for minggu in['minggu1', 'minggu2', 'minggu3','minggu4']:
    plt.plot(df['bulan'], df[minggu], marker='o', label=minggu)

plt.title("Tren pengeluaran mahasiswa perminggu")
plt.ylabel("Rupiah")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Diagram pie ([rpporsi pengeluaran per bulan)
plt.figure(figsize=(6,6))
plt.pie(df['Total_Pengeluaran'],labels=df['bulan'], startangle=90, autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.axis('equal')
plt.title("Proporsi pengeluaran mahasiswa perbulan")
plt.show()

```

```python
#regresi linear
#model Ai -> scikit, sklearn

from sklearn.linear_model import LinearRegression
df['bulan'] = np.arange(1,13)

X = df[['bulan']]
y = df['Total_Pengeluaran']

model = LinearRegression()
model.fit(X,y)

prediksi_13 = model.predict([[13]])[0]
print(f"Prediksi Pengeluaran Mahasiswa bulan Januari: {prediksi_13:,.0f}")
```
