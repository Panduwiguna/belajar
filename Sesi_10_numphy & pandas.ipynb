```python
import numpy as np
import pandas as pd

n_data = 10000


Hari = np.random.choice(['Senin','Selasa','Rabu','Kamis','Jumat','Sabtu','Minggu'], size=n_data)
Tanggal = np.random.randint(1,31, size=n_data)
Jam = np.random.randint(00.10,12.00, size=n_data)
Modal =np.random.choice(['1.000.000','500.000','700.000','600.000','400.000'], size = n_data)
Pembelian = np.random.randint(100.000, 500.000, size=n_data)
nama = np.random.choice(['susu', 'snack', 'beras', 'telur', 'gula'], size=n_data)
Transaksi = np.random.randint(300,400, size=n_data)
jumlah = np.random.randint(1,1000, size=n_data)
Penjualan = np.random.randint(100.000, 500.000, size=n_data)
Keuntungan = np.random.choice(['100.000','200.000','150.000','130.000'],size=n_data)

df = pd.DataFrame({
    'Hari' : Hari,
    'Tanggal' : Tanggal,
    'Jam' : Jam,
    'Modal' : Modal,
    'Pembelian' : Pembelian,
    'Barang' : nama,
    'Transaksi' :Transaksi,
    'Jumlah' : jumlah,
    'Penjualan' : Penjualan,
    'Keuntungan' : Keuntungan
})

df.to_csv('dataLatih.csv',index=False)

df.head ()# untuk mengetahui 5 baris pertama dari isi dataframe
```
